```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import scypy as sp
import matplotlib.pyplot as plt
```

```python
penguins = sns.load_dataset("penguins")
dow_johnes = sns.load_dataset("dowjones")
exercises = sns.load_dataset("exercise")
penguins.head()
```

## Histogram

```python
sns.histplot(
    data=penguins,
    x='bill_length_mm', 
    bins=20, 
    kde=True, 
    color='blue',
    hue="island"
)
plt.title('    ')
plt.xlabel('  ()')
plt.ylabel('')
plt.show()
```

## Piechart

```python
species_counts = penguins['species'].value_counts()

plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('   ')
plt.axis('equal')
species_counts.head()
```

## Scatterplot

```python
sns.scatterplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', hue='species', )

plt.title('     ')
plt.xlabel('  ()')
plt.ylabel('  ()')
plt.legend(title=' ')


plt.show()
```

## Scatterplot with mass

```python
sns.scatterplot(
    data=penguins, 
    x='bill_length_mm', 
    y='flipper_length_mm', 
    size='body_mass_g',
    hue='species', 
    palette='deep', 
    sizes=(20, 200), 
    legend=False
)

plt.title(' :        ')
plt.xlabel('  ()')
plt.ylabel('  ()')

plt.show()
```

## Linear plot

```python
sns.lmplot(
    data=penguins, 
    x='bill_length_mm', 
    y='flipper_length_mm',
    hue="species"
)

plt.title('     ')
plt.xlabel('  ()')
plt.ylabel('  ()')

plt.show()
```

## Linear plot with regression

```python
sns.lmplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', hue='species', order=2, ci=80)

plt.title('    ')
plt.xlabel('  ()')
plt.ylabel('  ()')

plt.show()
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
corr_matrix = diamonds[['carat', 'depth', 'table', 'price', 'x', 'y', 'z']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
```

 - annot=   True      .
 - cmap=     'coolwarm',        ,     .
 - vmin=  vmax=       .           ,      ( - ).

```python
plt.figure(figsize=(10, 8))
sns.heatmap(
    corr_matrix, 
    annot=True, 
    fmt=".2f", 
    linewidths=0.5, 
    cmap='coolwarm',
    square=True, 
    cbar_kws={'shrink': .8}, 
    vmax=1, 
    vmin=-1
)

plt.title('    ')
plt.show()
```

*           *fmt='.2f'*.
*      *linewidths=0.5*.
*      *square=True*.
* *cbar_kws={'shrink' : .8}*   ,      .

## Pair plots

```python
iris = sns.load_dataset('iris')
flights = sns.load_dataset("flights")
```

```python
sns.pairplot(
    iris, 
    hue='species', 
    palette='husl',
    diag_kind='kde', 
    markers=['o', 's', 'D'], 
    plot_kws={'alpha': 0.7})

plt.suptitle("     Iris", y=1.02)
plt.show()
```

## Matrix plots

```python
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")

plt.figure(figsize=(12, 8))
sns.heatmap(data=flights_pivot, annot=True, fmt="d", cmap="YlGnBu", linewidths=.5)

plt.title('      ')
plt.xlabel('')
plt.ylabel('')

plt.show()
```

# Styles

```python
tips = sns.load_dataset('tips')
```

* *whitegrid*        (   ).
* *darkgrid*        (,    ).
* *white*        ().
* *dark*        (    ).

```python
sns.set_style("whitegrid")

plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", data=tips, palette="pastel")
plt.title("whitegrid")
plt.show()
```

```python
sns.set_style("darkgrid")

plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", data=tips, palette="pastel")
plt.title("darkgrid")
plt.show()
```

```python
sns.set_style("white")

plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", data=tips, palette="pastel")
plt.title("white")
plt.show()
```

```python
sns.set_style("dark")

plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", data=tips, palette="pastel")
plt.title("dark")
plt.show()
```

      . Seaborn    ,   deep, muted, bright, pastel, dark,  colorblind,      .     **seaborn.set_palette('palette_name')**

   :
* **Qualitative**    (, deep, bright, pastel).
* **Sequential**    ,        (, Blues, Greens).
* **Diverging**         (, coolwarm, RdBu).
