```python
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
corr_matrix = diamonds[['carat', 'depth', 'table', 'price', 'x', 'y', 'z']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
```

 - annot= – при True показывает численное значение корреляции внутри квадрата.
 - cmap= – задает цветовую палитру 'coolwarm', где более холодные оттенки указывают на отрицаительную корреляцию, а теплые – на положительную.
 - vmin= и vmax= – указывают диапазон значений для цветовой шкалы. Без этих аргументов самое маленькое значение было бы самого холодного оттенка, что могло бы запутать нас (или кого-то другого).

```python
plt.figure(figsize=(10, 8))
sns.heatmap(
    corr_matrix, 
    annot=True, 
    fmt=".2f", 
    linewidths=0.5, 
    cmap='coolwarm',
    square=True, 
    cbar_kws={'shrink': .8}, 
    vmax=1, 
    vmin=-1
)

plt.title('Матрица корреляции для характеристик бриллиантов')
plt.show()
```

* Округлили значения корреляции до двух знаков после запятой с помощью *fmt='.2f'*.
* Добавили границы между ячейками посредством *linewidths=0.5*.
* Сделали ячейки квадратными при помощи *square=True*.
* *cbar_kws={'shrink' : .8}* уменьшает цветовую шкалу, чтобы она лучше вписывалась в график.

# Styles

```python
tips = sns.load_dataset('tips')
```

* *whitegrid* — стиль с белым фоном и сеткой (подходит для аналитических визуализаций).
* *darkgrid* — стиль с темным фоном и сеткой (универсальный, подходит для большинства графиков).
* *white* — стиль с белым фоном без сетки (минималистичный).
* *dark* — стиль с темным фоном без сетки (подходит для презентаций и отчетов).

```python
sns.set_style("whitegrid")

plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", data=tips, palette="pastel")
plt.title("whitegrid")
plt.show()
```

```python
sns.set_style("darkgrid")

plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", data=tips, palette="pastel")
plt.title("darkgrid")
plt.show()
```

```python
sns.set_style("white")

plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", data=tips, palette="pastel")
plt.title("white")
plt.show()
```

```python
sns.set_style("dark")

plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", data=tips, palette="pastel")
plt.title("dark")
plt.show()
```

Цветовая палитра — это основа выразительности графика. Seaborn предлагает несколько встроенных палитр, таких как deep, muted, bright, pastel, dark, и colorblind, а также позволяет создавать палитры вручную. Выставить палитру можно через **seaborn.set_palette('palette_name')**

Основные типы цветовых палитр:
* **Qualitative** — для категорий (например, deep, bright, pastel).
* **Sequential** — для непрерывных данных, где оттенки изменяются от светлого к темному (например, Blues, Greens).
* **Diverging** — для данных с положительными и отрицательными значениями (например, coolwarm, RdBu).

```python
# Построим на палитру для категорий
sns.set_palette("bright")

plt.figure(figsize=(10, 6))
sns.scatterplot(data=tips, x='tip', y="total_bill", hue="sex", alpha=0.7)
plt.legend(title="Пол")
plt.show()
```

```python
# Установим контекст для презентации
sns.set_context("talk")

plt.figure(figsize=(10, 6))
sns.boxplot(x="day", y="total_bill", data=tips, palette="muted")
plt.title("Сумма счета по дням недели (контекст презентации)")
plt.show()
```
