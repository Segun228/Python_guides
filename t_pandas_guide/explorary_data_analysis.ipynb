# Explorary data analysis

##### Getting and cleaning our data

```python
import pandas as pd
import matplotlib as ml


def get_df(path = "pharma.csv"):
    df = pd.read_csv(
        filepath_or_buffer=path,
        encoding="1251"
    )
    df = df.interpolate(
        method="linear"
    )
    return df
df = get_df()
df
```

### Building correlation mattrix

```python
import seaborn as sns
import matplotlib.pyplot as pplt

numeric_df = df.select_dtypes(include=["number"])
pplt.figure(figsize=(16, 6))
sns.heatmap(
    numeric_df.corr(),
    vmin=-1,
    vmax=1,
    annot=True,
    
)
```

#### Searching for connection between hour and slaes

```python
df.describe()

df["hour"] = pd.to_datetime(df.DR_Tim).dt.hour
data = df.groupby(["hour"])["DR_Kol"].agg(sum)
sns.boxplot(x = data)
```

```python

import seaborn as sns
import matplotlib.pyplot as plt


data_df = pd.DataFrame(data).reset_index()


sns.barplot(data=data_df, x="hour", y="DR_Kol")
plt.xlabel("Час")
plt.ylabel("Количество покупок")
plt.title("Покупки по часам")
plt.show()
```

## Groupping data

```python
df = get_df()
df_groupped = df.groupby(["DR_Dat", "DR_NChk"])[["DR_Kol", "DR_CRoz", "DR_CZak"]].agg(
    {
        "DR_Kol":"sum",
        "DR_CRoz":["sum", "max"],
        "DR_CZak": "sum"
    }
)
```

#### Finding income from every employee

```python
df_groupped = df.groupby(["DR_TabEmpl"])[["DR_Kol", "DR_CRoz", "DR_CZak"]].agg(
    {
        "DR_Kol":"sum",
        "DR_CRoz":["sum"],
        "DR_CZak": "sum"
    }
)
df_groupped["profit"] = df_groupped["DR_CRoz"] - df_groupped["DR_CZak"]
sns.barplot(data=df_groupped, x="DR_TabEmpl", y="profit", )


# first way 
df.groupby(["DR_TabEmpl", "DR_Dat"])[["DR_Kol", "DR_CRoz", "DR_CZak"]].agg(
    {
        "DR_Kol":"sum",
        "DR_CRoz":"sum",
        "DR_CZak": "sum"
    }
)

# second way
df.groupby(["DR_TabEmpl", "DR_Dat", "DR_CDrugs"])[["DR_Kol", "DR_CRoz", "DR_CZak"]].apply(lambda x: sum(x["roz"] - x["zak"]))
```
