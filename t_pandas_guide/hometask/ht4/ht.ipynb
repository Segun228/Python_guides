```python
import pandas as pd
import numpy as np
df = pd.read_csv(
    filepath_or_buffer="data.csv",
    parse_dates=["Date"]
)
df.head()

```

```python
df["Min_salary"].fillna(-1, inplace=True)
df["Max_salary"].fillna(-1, inplace=True)
df.dropna(inplace=True)


df.loc[(df["Min_salary"] == -1) & (df["Max_salary"] != -1), "Min_salary"] = df["Max_salary"]


df.loc[(df["Max_salary"] == -1) & (df["Min_salary"] != -1), "Max_salary"] = df["Min_salary"]


df = df[(df["Min_salary"] != -1) & (df["Max_salary"] != -1)]

df = df[df["Min_salary"] <= df["Max_salary"]]

df["Mean_salary"] = (df.Max_salary + df.Min_salary)/2

# Интервалы и метки
bins = [-1, 1, 3, 5, 7, 9]
labels = ['1', '2', '3', '4', '5']

# Применим группировку
df["experience_group"] = pd.cut(df["Experience"], bins=bins, labels=labels)
df.groupby(by="experience_group")['Mean_salary'].agg(['min', 'mean', 'median'])
```
