```python
import pandas as pd

df = pd.read_csv(
    filepath_or_buffer="data.csv",
    delimiter=";",
    skiprows=1
)
df["weekday"] = pd.to_datetime(df.date_joined).dt.day_of_week

df["weekday"] = df["weekday"].map(
    {
        0: "Понедельник",
        1: "Вторник",
        2: "Среда",
        3: "Четверг",
        4: "Пятница",
        5: "Суббота",
        6: "Воскресенье"
    }
)

df["date"] = pd.to_datetime(df.date_joined).dt.day

res = df.groupby(["weekday"], as_index=False).agg(
    reg_count = ("id", "count"),
    date_unique = ("date", "nunique"),
)
res
df
```

```python
import pandas as pd
df = pd.read_csv(
    filepath_or_buffer="data.csv",
    delimiter=";",
    skiprows=1
)
df = df[(df.id >= 94) & (~df.username.isin(['fnoxonsnbkbxsnb', 'Jonns21241412', 'fbxbnsn']))]
df["cohort"] = pd.to_datetime(df.date_joined).dt.to_period('M').astype(str)
res = df.groupby("cohort").agg(
    {
        "id":"count"
    }
).reset_index()
res.id = round(res.id / res.id.sum())
```

# Task 3

```python
import pandas as pd
import numpy as np
coderun = pd.read_csv(
    filepath_or_buffer="data2.csv",
    delimiter=",",
    skiprows=0
)
translator = { 
    3: "Python",
    2: "SQL"
}
coderun["cohort"] = pd.to_datetime(coderun.created_at).dt.strftime("%Y-%M")
coderun["language"] = coderun["language_id"].map(translator)
df = (
    coderun[coderun['cohort'] >= '2022-01']
    .groupby(["cohort", "language", "user_id"])["problem_id"]
    .agg("nunique")
    .groupby(["cohort", "language"])
    .agg(np.mean)
    .reset_index()
    .pivot_table(values='problem_id', columns=['language'], index=['cohort'])
)
```
