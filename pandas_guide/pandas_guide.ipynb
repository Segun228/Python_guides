## Работа с типами данных

### Изменение типа данных в одном из столбцов

```python
df.columns = ['ID', 'Name', "Age", "Department", "Salary", "City", "Date"]
df['Salary'] = df['Salary'].astype('float')
df.dtypes
```

### Форматирование столбца с датами

```python
df["Date"] = pd.to_datetime(df["Date"], format='mixed')
df["Date"] = df["Date"].dt.strftime("%d.%m.%Y")
df["Date"]
```

### Работа с датами

```python
pd.to_datetime(df["Date"], format='mixed').dt.day # Выведет день месяца
pd.to_datetime(df["Date"], format='mixed').dt.day_of_week # Выведет день недели
pd.to_datetime(df["Date"], format='mixed').dt.day_of_year # Выведет день года
```

### Получение уникальных значений

```python
df["Department"].unique()
```

### Замена одного из значений в столбце

```python
df["Department"] = df["Department"].astype('str').replace('Support', 'Support and Supply')
df["Department"]
```
