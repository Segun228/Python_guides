```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mplb
```

```python
chanels = pd.read_excel(
    io="data.xlsx",
    sheet_name="Данные по каналам"
)
#chanels["user_count"] = 1
chanels.head()
```

```python
costs = pd.read_excel(
    io="data.xlsx",
    sheet_name="Стоимость привлечения"
)
costs = costs.reset_index().rename(columns={
    "Unnamed: 0": "источник",
    "Стоимость привлечения, руб": "стоимость"
}).set_index("источник")
del costs["index"]
costs = costs.reset_index()

```

```python
groupped_users = chanels.groupby(by="Канал привлечения").agg(
    ["sum", "count"]
).reset_index()

total_table = pd.DataFrame()
total_table = groupped_users["LTV, руб."].reset_index()
total_table["index"] = groupped_users["Канал привлечения"]
total_table.rename({
    "index":"source"
})
```

```python

total = pd.merge(
    left=total_table,
    right=costs,
    left_on="index",
    right_on="источник",
    how="inner"
)
total["PM"] = (total["sum"] - total["стоимость"])
total["UPM"] = total["PM"] / total["count"]
total
```
